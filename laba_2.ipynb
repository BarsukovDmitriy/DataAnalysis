```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, precision_score
from sklearn.model_selection import GridSearchCV
```

```python
df = pd.read_csv('WineQT.csv')
df.head()
```

```python
# удалим столбец id - обычная индекация не несет в себе информативности
df = df.drop(['Id'], axis = 1)
df.head()
```

```python
# проверяем на пропуски
df.info()
# пропусков нет
```

```python
# разбивка на тестовую и тренировочную выборки
y = df['quality']
X = df.drop('quality', axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# создаем экземпляр и обучаем модель
classifier = GaussianNB()
classifier.fit(X_train, y_train)
```

```python
# считаем точность
y_pred = classifier.predict(X_test)
# расчет accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
# расчет precision
precision = precision_score(y_test, y_pred, average='micro')
print("Precision:", precision)
```

```python
X.corr()
```

```python
# подбор лучших параметров
param_grid = {
    'var_smoothing': [1e-10, 1e-8, 1e-7, 1e-6, 0.001, 0.0001, 0.000075]
}
grid_search = GridSearchCV(estimator=classifier, param_grid=param_grid, cv=5)
grid_search.fit(X_train, y_train)
best_params = grid_search.best_params_
best_score = grid_search.best_score_
print("Лучшие параметры:", best_params)
print("Лучшая оценка точности:", best_score)
```

```python
best_classifier = GaussianNB(var_smoothing=best_params['var_smoothing'])
best_classifier.fit(X_train, y_train)
y_pred = best_classifier.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
# расчет precision
precision = precision_score(y_test, y_pred, average='micro')
print("Precision:", precision)
```
