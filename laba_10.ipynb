```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score, precision_score
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv('WineQT.csv')
df.head()
```

```python
# удалим столбец id - обычная индекация не несет в себе информативности
df = df.drop(['Id'], axis = 1)
df.head()
```

```python
# проверяем на пропуски
df.info()
# пропусков нет
```

```python
# разбивка на тестовую и тренировочную выборки
y = df['quality']
X = df.drop('quality', axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# создаем экземпляр и обучаем модель
base_estimator = DecisionTreeClassifier(max_depth=1)  # Базовый оценщик (слабый классификатор)
n_estimators = 100  # Количество оценщиков (итераций)
adaboost = AdaBoostClassifier(base_estimator=base_estimator, n_estimators=n_estimators)
adaboost.fit(X_train, y_train)
```

```python
y_pred = adaboost.predict(X_test)
# расчет accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
# расчет precision
precision = precision_score(y_test, y_pred, average='micro')
print("Precision:", precision)
```

```python
# подберем оптимальные гиперпараметры
adaboost = AdaBoostClassifier(base_estimator=base_estimator)
param_grid = {
    'n_estimators': [50, 100, 200],
    'learning_rate': [0.1, 0.5, 1.0]
}
grid_search = GridSearchCV(adaboost, param_grid, scoring='accuracy', cv=5)
grid_search.fit(X_train, y_train)
best_model = grid_search.best_estimator_
best_params = grid_search.best_params_
```

```python
y_pred = best_model.predict(X_test)
# расчет accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
# расчет precision
precision = precision_score(y_test, y_pred, average='micro')
print("Precision:", precision)
```
